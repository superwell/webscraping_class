```python
import requests
from bs4 import BeautifulSoup
```

```python
result = requests.get("https://hk.finance.yahoo.com/quote/0001.HK/history?p=0001.HK&.tsrc=fin-srch")
```

```python
result.status_code
```

```python
result.headers
```

```python
content = result.content
```

```python
content
```

```python
soup = BeautifulSoup(content)
```

```python
soup.findAll('h1')
```

```python
title = soup.findAll('h1')[0]
```

```python
title.text
```
